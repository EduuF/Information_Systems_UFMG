# Teste de hipótese/IC 1

Beth recebeu de amigos experientes uma informação de que caso estivesse com orçamento pequeno, jogadores com chutes potentes são melhores contratações do que aqueles com precisão de chute muito alta pois o salário ESPERADO deles é menor.
Beth decidiu então fazer um intervalo de confiança (95%) para os 100 melhores jogadores em cada uma das características e compará-los para confirmar se a informação se confirmava.


## Função de IC

```python
def calcula_IC_salario(df, feature, top_x): #feature = feature a ser comparada c/ salario
                                            #top_x = seleciona os x jogadores com maior pontuação na feature
        
    df2 = df[['Wage' ,feature]] #Seleciona colunas de interesse
    df2 = df.nlargest(top_x,[feature]) #Os 100 jogadores com chutes mais potentes
    salario = df2['Wage']
    s = np.std(salario, ddof=1) ##1 grau de liberdade
    s_sobre_raiz_n = (s / np.sqrt(len(salario)))
    media = np.mean(salario)
    inferior = media-(1.96 * s_sobre_raiz_n)
    superior = media+(1.96 * s_sobre_raiz_n)
    return [ inferior, superior ]
```

## Qual o IC dos salários dos 100 jogadores com chutes mais potentes?

```python
#Selecionando os 100 jogadores com chutes mais potentes
df_2 = df[['Wage' ,'ShotPower']] #Seleciona colunas de interesse
df_2 = df.nlargest(100,['ShotPower'])

#Plotando os dados
plt.scatter(df_2['ShotPower'], df_2['Wage'], Alpha=0.2)
plt.axvline(df_2['ShotPower'].mean(), color='r', linewidth=1)
plt.axhline(df_2['Wage'].mean(), color='r', linewidth=1)
plt.title('Relação entre Salário vs Força do chute com dados normalizados')
plt.xlabel("Força do chute")
plt.ylabel("Salário")
plt.show()
```

### Fazendo o IC 

```python
shotPower_Wage_IC = calcula_IC_salario(df, 'ShotPower', 100)
shotPower_Wage_IC
```

Intervalo de confiança (95%) para o salário dos 100 jogadores com os chutes mais potentes (em €1000)

## Qual o IC dos salários dos 100 jogadores com maior precisão em chutes de longa distância?

```python
#Selecionando os 100 jogadores com chutes mais potentes
df_3 = df[['Wage' ,'LongShots']] #Seleciona colunas de interesse
df_3 = df.nlargest(100,['LongShots'])

#Plotando os dados
plt.scatter(df_3['LongShots'], df_3['Wage'], Alpha=0.2)
plt.axvline(df_3['LongShots'].mean(), color='r', linewidth=1)
plt.axhline(df_3['Wage'].mean(), color='r', linewidth=1)
plt.title('Relação entre Salário vs Precisão de chutes a longa distância com dados normalizados')
plt.xlabel("LongShots")
plt.ylabel("Salário")
plt.show()
```

### Fazendo o IC

```python
longShots_Wage_IC = calcula_IC_salario(df, 'LongShots', 100)
longShots_Wage_IC
```

Intervalo de confiança (95%) para o salário dos 100 jogadores com maior precisão de chute (em €1000)

# Regressão

Beth sabe que para seu time ir bem no longo prazo precisa contratar jogadores para se desenvolverem dentro do clube (se desenvolver = ter um alto ‘Overall’), por isso pediu a um cientista de dados um modelo que ela pudesse utilizar para tentar prever novas promessas do futebol baseado em seus status.

### Feature selection

```python
#Tirando features que de cara sei que não vão ser úteis para o modelo
X = df.drop(['Name', 'Age', 'Wage' ,'Nationality', 'Club', 'Preferred Foot', 'Potential', 'Jersey Number'],1)
```

```python
#Continuando o feature selection
#Função para plotar a relação entre as features
def plot_corr(df, size):
    corr = df.corr()
    fig, ax = plt.subplots(figsize = (size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)

plot_corr(X, len(X.columns))
```

Ficou claro nessa gráfico que as features relacionadas aos goleiros destoa muito das outras, o que nos leva a entender que devemos ter modelos diferentes para jogadores de linha e goleiros

```python
#Separando DF dos goleiros dos demais jogadores

#X é o DF dos jogadores de linha
X = X.loc[X['Position'] != 'GK']
```

```python
#Tirando as features relativas aos goleiros
X = X.drop(['GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes', 'Position'],1)
```

```python
#Mostrando numericamente a relação entre as features do DF
X.corr()
```

```python
#Tirando as features cujo o VALOR ABSOLUTO da correlação com 'Overall' seja menor que 0.47
X = X.drop(['Weak Foot', 'Acceleration', 'SprintSpeed' ,'Agility', 'Balance', 'Jumping', 'Strength', 'Interceptions'],1)
X = X.drop(['Marking', 'StandingTackle', 'SlidingTackle', 'Positioning', 'Penalties', 'HeadingAccuracy'],1)
X = X.drop(['Volleys', 'FKAccuracy', 'Stamina', 'Aggression'],1)
```

```python
#Plotando novo gráfico de correlações, com as features atualizadas
plot_corr(X, len(X.columns))
```

### Lidando com os dados

```python
#Setando a feature 'overall' como variável alvo para a regressão e tirando features não numéricas de X
y = X['Overall']
X = X.drop(['Overall'],1)
```

```python
#Normalizando os dados
X_mean = X.mean()
X_std = X.std(ddof=1)
y_mean = y.mean()
y_std = y.std(ddof=1)

X = X - X_mean
y = y - y_mean
X = X/X_std
y = y/y_std

#Adicionando intercept
X.insert(0, 'intercept', 1.0)

#Pegando os valores
X = X.values
y = y.values
```

### Lidando com Missing Values Ocultos = 0

```python
#Lidando com missing ocultos 0 substituindo pela média
from sklearn.impute import SimpleImputer

missing_0 = SimpleImputer(missing_values = 0, strategy = "mean")
X = missing_0.fit_transform(X)
```

### Separando dados e criando o moledo

```python
from sklearn.model_selection import train_test_split

#Separando os dados para treino(70%) e teste(30%)
train_test_size = 0.3

#Separando
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=train_test_size)
```

```python
from sklearn.linear_model import LinearRegression

#Criando o modelo
lr_classifier = LinearRegression().fit(X_treino, y_treino.ravel())
```

### Avaliando o modelo com Coeficiente de determinação (R2)

```python
from sklearn.metrics import r2_score
```

```python
lr_predict_treino = lr_classifier.predict(X_treino)
lr_predict_teste = lr_classifier.predict(X_teste)
```

```python
r2_treino = r2_score(y_treino, lr_predict_treino)
r2_teste = r2_score(y_teste, lr_predict_teste)
```

```python
print("Coeficiente de determinação para os dados de treino: {0:.5f}".format(r2_treino))
print("Coeficiente de determinação para os dados de teste: {0:.5f}".format(r2_teste))
```

Podemos notar que os valores estão bem próximos, portanto o modelo é válido
